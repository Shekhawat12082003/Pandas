# **Creating Dataframes**

```python
# Using lists
Students_Data = [
    [100,80,10],
    [98,70,8],
    [80,58,2]
]
pd.DataFrame(Students_Data,columns=['iq','marks','dataframes'])
```

```python
# Using Dicts
Students_Data = {
    'iq':[100,98,80],
    'marks':[80,70,58],
    'dataframes':[10,8,2]
}
pd.DataFrame(Students_Data)
```
